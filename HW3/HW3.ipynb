```python
import re

def clean_ria_intro(text: str) -> str:
    # Удаляем всё от начала строки до конца выражения "РИА <любое_слово>" (например, "РИА Новости", "РИА Недвижимость" и т.д.)
    cleaned = re.sub(r"^.*?РИА\s+\S+\s*[-—]*\s*", "", str(text))
    return cleaned.strip()

data_ria["text"] = data_ria["content"].apply(clean_ria_intro)
```
